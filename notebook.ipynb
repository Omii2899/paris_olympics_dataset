```python

import os
import pandas as pd
import ast
```

```python
#               kaggle datasets download -d piterfm/paris-2024-olympic-summer-games
```

```python
filepath = "./paris-2024-olympic-summer-games"
```

```python
dataset_dict = dict()
for x in os.listdir(filepath):
    dataset_path = filepath + '/' + x
    dataset_name = x.split(".")[0]
    dataset = pd.read_csv(dataset_path)
    dataset_dict[dataset_name] = dataset
```

```python
for x in dataset_dict:
    print(x, dataset_dict[x].columns)
```

## Venues

```python
x = dataset_dict['venues']
```

```python


def convert_to_csv(cell):
    try:
        # Evaluate the string as a literal list
        lst = ast.literal_eval(cell)
        # Join the elements, preserving the words
        return ', '.join(lst)
    except:
        # If evaluation fails, return the original string
        return cell

# Assuming your DataFrame is named 'df' and the column with list-like strings is named 'list_column'
x['sports'] = x['sports'].apply(convert_to_csv)
```

```python
x = x.drop(columns=['date_start', 'date_end', 'url'])
```

## Schedules

```python
x = dataset_dict['schedules']
```

```python
def remove_numbers(string):
    new_string = str()
    for x in string:
        if not x.isdigit():
            new_string += x.strip()
    return new_string
```

```python
x['venue_code'] = x['venue_code'].apply(lambda x: remove_numbers(x) if isinstance(x, str) else x)
```

```python
x['venue'] = x['venue'].apply(lambda x: remove_numbers(x) if isinstance(x, str) else x)
```
